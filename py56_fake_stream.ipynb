```
import datetime
import logging
import time
from random import randint, random

import numpy as np
from pyspark.sql import SparkSession

logging.basicConfig(level=logging.INFO)
logging.info("Start stream")
```

```
def build_consumption_event():
    consumer = np.random.choice(['Watson', 'Skinner',
                                 "Zimbardo", "Pavlow"],
                                replace=False, p=[0.2, 0.3, 0.35, 0.15])
    consumer = str(consumer)
    consum = {'BeerID': randint(1, 30),
              'DrinkingTime': datetime.datetime.now(),
              'DrinkingVolume': random() * 0.3 * len(consumer) / 6 + 1,
              'Consumer': consumer}
    return consum
```

```
spark.catalog.setCurrentCatalog('spark_training_stese')
spark.conf.set('spark.databricks.delta.properties.defaults.enableChangeDataFeed', 'true')
spark.sql("drop table if exists beer_consumptions_streaming")
for y in range(10000):
    content = [build_consumption_event() for i in range(20)]
    print(content)
    df = spark.createDataFrame(content)
    df.write.saveAsTable('beer_consumptions_streaming', mode='append')
    logging.info(f"Added streaming batch")
    time.sleep(10)
```
